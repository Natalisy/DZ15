```python
envCH = pd.read_csv('Environment_Temperature_change.csv', encoding='cp1250')
envCH
```

```python
# Что интересует в первую очередь? Посмотреть изменения температуры в России. 
# И... Получаем кучу NaN-ов. Хотя полное отсутствие данных до распада СССР закономерно. Также как и 
# пустые строки стандартного отклонения

df_Russia = envCH[envCH['Area'] =='Russian Federation']
df_Russia.head()
```

```python
# Но все равно попробуем выцепить что-то полезное. Сначала возьмем только интересующие нас столбцы

df_Russia_short = df_Russia.drop(['Area Code', 'Area', 'Months Code', 'Element Code', 'Unit'], axis = 1) 
df_Russia_short.head()
```

```python
df_Russia_actual_columns = df_Russia_short.drop(df_Russia_short.ix[:, 'Y1961':'Y1991'].columns, axis = 1) 
df_Russia_actual_columns.head()
```

```python
# Итого получился совершенно нерепрезентативный датафрейм с минимальным объемом данных)
# Но для цели поиграться с визуализацией, надеюсь, подойдет...

Russia4work = df_Russia_actual_columns[df_Russia_actual_columns['Element'] == 'Temperature change']
Russia4work
```

```python
Russia4work = Russia4work.drop(['Element'], axis = 1)
Russia4work
```

```python
Russia4work.set_index('Months', inplace=True)
```

```python
Russia4work
```

```python
Russia4work.index.names = [None]
Russia4work
```

```python
# Многовато будет...

# Плюс упустили одну важную деталь. Среди строк по каждой стране присутствует и разбиение по сезонам,
# и "Метеорологический год". Что при выводе в общем графике с отдельными месяцами существенно искажает
# картину. С другой стороны, интересно посмотреть общую картину и на таких данных.

Russia4work.plot()
```

```python
# Вот интересное наблюдение: ValueError: When stacked is True, each column must be either all positive or negative.
# Y1992 contains both positive and negative values

# То есть данный вид графика для датасетов с температурой (или другими подобными данными) не подойдет.
# Либо нужно как-то нормировать

Russia4work.plot(kind='area')
```

```python
# Разобьем наш, и без того маленький датасет, на 3 отдельных)
# Возьмем отдельно строку "Метеорологический год"

meteo_year = Russia4work.loc[['Meteorological year']]
meteo_year
```

```python
# Оставлю здесь этот блок. Второй раз сталкиваюсь с ошибками типа этой. Другой вариант был:
# AttributeError: 'Line2D' object has no property 'xlabel'. Подскажите, пожалуйста, с чем это может
# быть связано и как решать такие проблемы?

ax = meteo_year.plot(title= 'Meteorological year', kind='hist', xlabel='Temperature change')
ax.legend(loc='lower center',ncol = 10)
```

```python
# И вот здесь наглядно выясняется интересная деталь - количесвто цветов, которые заложены в plot() по
# дефолту очень маленькое. В итоге показатели для разных периодов дублируются в цветах аж по 3 раза...

ax = meteo_year.plot(title= 'Meteorological year', kind='hist', stacked=True)
ax.legend(loc='lower center',ncol = 10)
# plt.yticks(np.arange(-0.5,5,0.5))
```

```python
# Возьмем отдельно по месяцам
months = Russia4work.iloc[0:12]
months
```

```python
# В любом случае видим одну и ту же картину: наибольший разброс по изменениям температуры наблюдается
# в осенне-зимний период. Период с мая по сентябрь показывает на удивление стабильные показатели год 
# от года.

# Учитывая, что речь идет о глобальном потеплении, возможно, это как раз ожидаемый результат. 

months_ax = months.plot(title='Разброс изменений температуры по месяцам в течении 10 лет')
months_ax.legend(loc='lower center',ncol = 10)
```

```python
# Пока непонятно, как правильно пользоваться этим графиком...

months_ax = months.plot(kind='scatter')
months_ax.legend(loc='lower center',ncol = 10)
```

```python
# неожиданный список None-ов... О_о А как быть в таких случаях? 
# Можно ли этот список как-то убрать из вывода?

# Среди прочего бросается в глаза цикличность (волнообразность) в ящиках. Как по разбросу значений 
# в рамках каждого года, так и по самой интенсивности колебаний температуры. Насколько мне известно, 
# подобная цикличность является абсолютно нормальной для климата. Поэтому наш сетик из 1-го десятилетия 
# для таких глобальных показателей совсем неинформативен

months_ax = months.plot(kind='box',
                        title='Разброс изменений температуры по месяцам в течении 10 лет')
months_ax.legend(loc='lower center',ncol = 10)
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha="right")
```

```python
months_ax = months.plot(kind='hist',
                        title='Разброс изменений температуры по месяцам в течении 10 лет',
                       stacked=True)
months_ax.legend(loc='lower center',ncol = 10)
```

```python
months_ax = months.plot(kind='bar',
                        title='Разброс изменений температуры по месяцам в течении 10 лет',
                       stacked=True)
months_ax.legend(loc='upper center',ncol = 14)
```
